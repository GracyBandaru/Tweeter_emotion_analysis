```python
import pandas as pd
from transformers import pipeline

# Load the CSV file
data_path = r'/content/data.csv'
data = pd.read_csv(data_path)
```

```python
if 'Tweets' not in data.columns:
    raise ValueError("The file must contain a 'text' column for analysis.")

emotion_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=None)

```

```python

print("Analyzing emotions...")
data['emotions'] = data['Tweets'].apply(lambda x: emotion_classifier(x))


output_path = r'/content/data.csv'
data.to_csv(output_path, index=False)
```

```python
print(f"Emotion analysis completed. Results saved to {output_path}")

def check_emotion(input_text):
    result = emotion_classifier(input_text)
    return result

if __name__ == "__main__":
    user_input = input("Enter a text to analyze its emotion: ")
    emotion_result = check_emotion(user_input)
    print("Emotion analysis result:", emotion_result)
```
